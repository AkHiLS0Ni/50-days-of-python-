```python
num = int(input("Enter number: "))

if num >= 0:
    if num == 0:
        print("Number is Zero")
    else:
        print("Number is Positive")
else:
    print("Number is Negative")
```

```python
age = int(input("Enter age: "))
citizen = input("Are you a citizen (yes/no): ")

if age >= 18:
    if citizen == "yes":
        print("You can vote")
    else:
        print("You must be a citizen to vote")
else:
    print("You are not eligible to vote")
```

```python
num = int(input("Enter number: "))

if num % 2 == 0:
    if num > 10:
        print("Even and greater than 10")
    else:
        print("Even but less than or equal to 10")
else:
    print("Number is odd")
```

```python
marks = int(input("Enter marks: "))

if marks >= 50:
    if marks >= 90:
        print("Grade A")
    elif marks >= 75:
        print("Grade B")
    else:
        print("Grade C")
else:
    print("Fail")
```

```python
year = int(input("Enter year: "))

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("Leap Year")
        else:
            print("Not Leap Year")
    else:
        print("Leap Year")
else:
    print("Not Leap Year")
```

```python
username = input("Enter username: ")
password = input("Enter password: ")

if username == "admin":
    if password == "1234":
        print("Login successful")
    else:
        print("Wrong password")
else:
    print("User not found")
```

```python
num = int(input("Enter number: "))

if num % 5 == 0:
    if num % 3 == 0:
        print("Divisible by 5 and 3")
    else:
        print("Divisible by 5 but not by 3")
else:
    print("Not divisible by 5")
```

```python
salary = int(input("Enter salary: "))
experience = int(input("Enter years of experience: "))

if salary > 50000:
    if experience > 5:
        print("Eligible for bonus")
    else:
        print("Not enough experience")
else:
    print("Salary too low for bonus")
```
