```python
numbers = [10, 20, 30, 40, 50]

total = 0

for num in numbers:
    total += num

average = total / len(numbers)

print("Average:", average)
```

```python
numbers = [1, 2, 3, 4]

product = 1

for num in numbers:
    product *= num

print("Product:", product)
```

```python
list1 = [1, 2, 3, 4]
list2 = [3, 4, 5, 6]

common = []

for num in list1:
    if num in list2:
        common.append(num)

print("Common elements:", common)
```

```python
numbers = [1, 2, 2, 3, 4, 4, 5]

unique = []

for num in numbers:
    if num not in unique:
        unique.append(num)

print("List without duplicates:", unique)
```

```python
numbers = [10, 25, 7, 30, 18]

numbers.sort()

print("Second largest:", numbers[-2])
```

```python
numbers = [1, 2, 3, 4, 5, 6]

even = 0
odd = 0

for num in numbers:
    if num % 2 == 0:
        even += 1
    else:
        odd += 1

print("Even:", even)
print("Odd:", odd)
```

```python
numbers = [1, 2, 3, 4]

reversed_list = []

for i in range(len(numbers)-1, -1, -1):
    reversed_list.append(numbers[i])

print("Reversed list:", reversed_list)
```

```python
numbers = [5, 12, 8, 20, 3]

largest = numbers[0]

for num in numbers:
    if num > largest:
        largest = num

print("Largest:", largest)
```

```python
numbers = [7, 2, 10, 5]

smallest = numbers[0]

for num in numbers:
    if num < smallest:
        smallest = num

print("Smallest:", smallest)
```

```python
numbers = [1, 2, 3, 4, 5]

first = numbers[0]

for i in range(1, len(numbers)):
    numbers[i-1] = numbers[i]

numbers[-1] = first

print("Rotated list:", numbers)
```
