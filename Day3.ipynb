| Operator | Meaning          |
| -------- | ---------------- |
| `==`     | Equal            |
| `!=`     | Not equal        |
| `>`      | Greater than     |
| `<`      | Less than        |
| `>=`     | Greater or equal |
| `<=`     | Less or equal    |


```python
x = 5
if x != 10:
    print("Not 10")
```

| Operator | Meaning           |
| -------- | ----------------- |
| `and`    | Both must be True |
| `or`     | At least one True |
| `not`    | Reverse condition |


```python
age = 20
citizen = True

if age >= 18 and citizen:
    print("You can vote")
```

```python
age = 20
if age >= 18:
    if age >= 21:
        print("You can drink and vote")
    else:
        print("You can vote but not drink")
```

```python
if 5 > 2: print("Five is greater")
```

value_if_true if condition else value_if_false

```python
age = 17
result = "Adult" if age >= 18 else "Minor"
print(result)
```
