# Day 9 Python Plan – Strings

```python
text = "python"
print(len(text))
```

```python
text = "python"

print(text[0])   # p
print(text[3])   # h
```

```python
text = "python"

print(text[0:3])   # pyt
print(text[2:5])   # tho
```

```python
text = "python"

print(text[::-1])
```

```python
text = "python"

print(text.upper())
print(text.capitalize())
```

```python
text = "python"

for ch in text:
    print(ch)
```

```python
text = "programming"

count = 0

for ch in text:
    if ch in "aeiou":
        count += 1

print("Vowels:", count)
```

```python
text = "madam"

if text == text[::-1]:
    print("Palindrome")
else:
    print("Not palindrome")
```

```python
text = "programming"

freq = {}

for ch in text:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1

print(freq)
```
