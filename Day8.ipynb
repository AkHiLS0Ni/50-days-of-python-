```python
lst = [1,2,2,3,4,4,5]

unique = []

for i in lst:
    if i not in unique:
        unique.append(i)

print(unique)
```

```python
a = [1,2,3,4]
b = [3,4,5,6]

common = []

for i in a:
    if i in b:
        common.append(i)

print(common)
```

```python
lst = [[1,2],[3,4],[5,6]]

flat = []

for sub in lst:
    for i in sub:
        flat.append(i)

print(flat)
```

```python
lst = [1,2,2,3,3,3]

for i in set(lst):
    print(i, "->", lst.count(i))
```

```python
lst = [1,2,3,4,5]

k = 2

result = lst[k:] + lst[:k]

print(result)
```

```python
lst = [1,2,3,4,5,6]

even = []
odd = []

for i in lst:
    if i % 2 == 0:
        even.append(i)
    else:
        odd.append(i)

print("Even:", even)
print("Odd:", odd)
```

```python
lst = [2,3,4]

result = 1

for i in lst:
    result *= i

print(result)
```

```python
lst = [1,2,3,4,5]

target = 5

for i in range(len(lst)):
    for j in range(i+1,len(lst)):
        if lst[i] + lst[j] == target:
            print(lst[i], lst[j])
```

```python
lst = [3,-1,5,-7,8,-2]

result = []

for i in lst:
    if i > 0:
        result.append(i)

print(result)
```
