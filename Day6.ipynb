```python
n = int(input("Enter a number: "))

sum = 0
for i in range(1, n+1):
    sum += i

print("Sum:", sum)
```

```python
num = int(input("Enter number: "))
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10

print("Reversed Number:", rev)
```

```python
num = int(input("Enter number: "))
count = 0

while num > 0:
    num //= 10
    count += 1

print("Total digits:", count)
```

```python
n = int(input("Enter number: "))
fact = 1

for i in range(1, n+1):
    fact *= i

print("Factorial:", fact)
```

```python
n = int(input("Enter number of terms: "))

a, b = 0, 1

for i in range(n):
    print(a, end=" ")
    a, b = b, a+b
```

```python
num = int(input("Enter number: "))
flag = True

for i in range(2, num):
    if num % i == 0:
        flag = False
        break

if flag:
    print("Prime number")
else:
    print("Not prime")
```

```python
num = int(input("Enter number: "))
sum = 0

while num > 0:
    digit = num % 10
    sum += digit
    num //= 10

print("Sum of digits:", sum)
```

```python
num = int(input("Enter number: "))
temp = num
sum = 0

while num > 0:
    digit = num % 10
    sum += digit ** 3
    num //= 10

if sum == temp:
    print("Armstrong number")
else:
    print("Not Armstrong")
```
